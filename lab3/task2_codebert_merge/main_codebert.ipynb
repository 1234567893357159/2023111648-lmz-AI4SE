## 4b. CodeBERT 微调训练

使用 `microsoft/codebert-base`，在训练集上微调，验证集上选最佳模型。

```python
# 跳过开关：True=加载已有模型，False=重新训练
SKIP_TRAIN = False

from codebert_merge import CodeBERTMergeTrainer

trainer = CodeBERTMergeTrainer()

if SKIP_TRAIN and os.path.exists(os.path.join(config.CODEBERT_MODEL_DIR, "config.json")):
    trainer.load()
else:
    trainer.train()
```

## 4c. 测试集评估

```python
result = trainer.evaluate()

y_true = result["labels"]
y_pred = result["predictions"]
y_proba = result["probabilities"]
```

## 4d. 可视化

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, roc_curve, auc
import seaborn as sns

os.makedirs(config.PLOTS_DIR, exist_ok=True)

# 混淆矩阵
cm = confusion_matrix(y_true, y_pred)
fig, ax = plt.subplots(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=ax,
            xticklabels=["Not Merged", "Merged"],
            yticklabels=["Not Merged", "Merged"])
ax.set_xlabel("Predicted")
ax.set_ylabel("True")
ax.set_title("CodeBERT - Confusion Matrix")
path = os.path.join(config.PLOTS_DIR, "confusion_matrix_CodeBERT.png")
fig.savefig(path, dpi=150, bbox_inches="tight")
plt.close(fig)
print(f"混淆矩阵已保存: {path}")

# ROC 曲线
fpr, tpr, _ = roc_curve(y_true, y_proba)
roc_auc = auc(fpr, tpr)
fig, ax = plt.subplots(figsize=(6, 5))
ax.plot(fpr, tpr, color="#2ecc71", lw=2, label=f"CodeBERT (AUC={roc_auc:.4f})")
ax.plot([0, 1], [0, 1], color="gray", lw=1, linestyle="--")
ax.set_xlabel("False Positive Rate")
ax.set_ylabel("True Positive Rate")
ax.set_title("CodeBERT - ROC Curve")
ax.legend(loc="lower right")
ax.grid(True, alpha=0.3)
path = os.path.join(config.PLOTS_DIR, "roc_curve_CodeBERT.png")
fig.savefig(path, dpi=150, bbox_inches="tight")
plt.close(fig)
print(f"ROC 曲线已保存: {path}")
```
